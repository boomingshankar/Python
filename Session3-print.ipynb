```python
a=500
b=300
addition=a+b
substraction=a-b
multiplication=a*b
print(f'Addition of {a} and {b} is {addition}')
print(f'Substraction of {a} and {b} is {substraction}')
print(f"Multiplication of {a} and {b} is {multiplication}")
```
