```python
import random
num = random.randint(1,100)
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
dis=eval(input('enter the km to travel'))
govt = eval(input('free ride distance form gov'))
if dis>govt:
    charge = eval(input('enter the cost of travel'))
    charge_on = dis-25
    print(f'the amount need to pay for additional distance {charge_on}km travel is {charge_on*charge}')
else:
    print('enjoy your free ride by the gov..')
```

```python
numb = eval(input('Please enter number :'))
if numb >= 0:
    if numb == 0:
        print(f'The number {numb} is Zero')
    else:
        print(f'Number {numb} is Positive')      
else:
    print(f'Number {numb} is negative')
```
