```python
shape = input("Enter the shape: ")
if shape == 'Triangle':
    try:
        radius = eval(input('Enter the radius: '))
        pi = 3.142
        area = pi*radius*radius
        print(f"The area of the {shape} is: {area}")
    except Exception as e:
        print(f"The area cannot br calculated due to this error: {e}")
    finally:
        print("All the best for coding")
elif shape == 'Rectangle':
    try:
        length = eval(input("What is the length: "))
        bredth = eval(input("What is the bredth: "))
        r_area = length*bredth
        print(f"The area of the {shape} is: {r_area}")
    except Exception as a:
        print(f"The rectangle cannot be calculated due to this error: {a}")
    finally:
        print("I always execute")
else:
    print("Invalid shape requested.")
```
