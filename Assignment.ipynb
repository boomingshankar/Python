```python
# wap to calculate area of rectangle
length = input("Please enter length ")
width = input("Please enter width ")
area = float(length) * float(width)
print(f"The area of rectange for given length {length} and width {width} is: {area}")
```

```python
#wap to caluclate simple intrest
principle_amount = input("Enter the priciple amount: ")
rate = input("Enter the rate: ")
time = input("Enter the time in years: ")
pa = float(principle_amount)
rt = float(rate)
tm = float(time)
simple_intrest = ( pa * rt * tm ) / 100
print(f"The simple intrest is: {simple_intrest}")
```

```python
#Write a program to convert temperature from celcius to farenheit
celius = input("Please enter temprature in celcius ")
farenheit = (float(celius) * 9/5) +32
print(f"The temprature in Farenheit is: {farenheit}")
```

```python
#Wap to convert min to sec
min = input("Enter the minutes: ")
sec = int(min) * 60
print(f"The conversion on minutes {min} into seconds: {sec}")
```

```python
#Wap to calulate volume of cube
side = input("Enter the side of cube ")
sd = float(side)
volume = sd * sd * sd
print(f"The total volume for side {sd} is: {volume}")
```

```python
#WAP to find the perimeter of rectangle Formula: perimeter = 2 * (length + width)
length = eval(input("Enter the length: "))
width = eval(input("Enter the width: "))
perimeter = 2 * (length + width)
print(f"The Perimeter of Rectangle is: {perimeter}")
```
