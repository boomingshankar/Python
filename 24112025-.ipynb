```python
percentage = eval(input("Whats your percentage??"))
if percentage >= 75:
    print("A grade")
elif percentage >= 60 :
    print("B grade")
elif percentage >= 45 :
    print("C grade")
elif percentage >= 30 :
    print("D grade")
else:
    print("You fail")
```
